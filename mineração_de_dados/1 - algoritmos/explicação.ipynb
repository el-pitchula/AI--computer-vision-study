### Conceitos básicos de cálculo
Funções e derivadas

Uma função é uma relação matemática que associa um conjunto de entradas a um conjunto de saídas. Em machine learning, frequentemente trabalhamos com funções que descrevem o comportamento dos modelos e como eles se ajustam aos dados.

A derivada de uma função é uma medida de sua taxa de variação em relação à sua variável independente. Em outras palavras, a derivada nos mostra a inclinação da função em um determinado ponto. Derivadas são especialmente úteis em algoritmos de machine learning, pois nos ajudam a otimizar as funções de custo e a encontrar os parâmetros ideais para nossos modelos.

Exemplo: Função quadrática e sua derivada

Considere a função quadrática f(x) = x^2. Vamos plotar a função e sua derivada f'(x) = 2x usando a biblioteca Matplotlib:

```python
# Importa as bibliotecas numpy e matplotlib.pyplot
import numpy as np
import matplotlib.pyplot as plt

# Define a função quadrática f(x) = x^2
def f(x):
    return x ** 2

# Define a função que calcula a derivada da função quadrática f(x)
def df(x):
    return 2 * x

# Cria um array de 1000 pontos no intervalo [-10, 10] para o eixo x
x = np.linspace(-50, 50, 1000)

# Calcula os valores da função quadrática e sua derivada para cada ponto no eixo x
y = f(x)
dy = df(x)

# Cria um gráfico com os valores de x no eixo x e os valores de y no eixo y
plt.plot(x, y, label="f(x) = x^2")
# Cria um gráfico com os valores de x no eixo x e os valores de dy no eixo y
plt.plot(x, dy, label="f'(x) = 2x")

# Adiciona um rótulo para o eixo x
plt.xlabel("x")
# Adiciona um rótulo para o eixo y
plt.ylabel("y")
# Adiciona uma legenda ao gráfico
plt.legend()
# Adiciona um título ao gráfico
plt.title("Função quadrática e sua derivada")
# Adiciona uma grade ao gráfico
plt.grid()

# Mostra o gráfico na tela
plt.show()
```

### Gradientes e otimização

Em algoritmos de machine learning, muitas vezes trabalhamos com funções multivariáveis, ou seja, funções que possuem múltiplas variáveis independentes. O gradiente de uma função multivariável é um vetor que contém as derivadas parciais em relação a cada uma das variáveis independentes. O gradiente nos indica a direção de maior aumento da função, e seu oposto, a direção de maior descida.

A otimização é o processo de encontrar os valores das variáveis independentes que minimizam ou maximizam uma função. Em machine learning, geralmente queremos minimizar a função de custo para ajustar nosso modelo aos dados. Uma otimização é o método do gradiente descendente, no qual iterativamente atualizamos os parâmetros do modelo na direção oposta ao gradiente.

### Integração e áreas sob curvas

A integração é o processo inverso da diferenciação e nos permite calcular a área sob a curva de uma função. Em machine learning, a integração é usada, por exemplo, para calcular probabilidades acumuladas em distribuições de probabilidade.

Exemplo: Área sob a curva de uma função quadrática

Vamos calcular a área sob a curva da função quadrática f(x) = x^2 no intervalo [0, 2] usando a função quad da biblioteca SciPy:

```python
# Importar a função "quad" do módulo "integrate" da biblioteca "scipy"
from scipy.integrate import quad

# Definir a função que será integrada
def f(x):
    return x ** 2

# Calcular a integral da função no intervalo [0, 2] usando a função "quad"
# A função "quad" retorna o valor da integral e uma estimativa do erro
# Como estamos interessados apenas no valor da integral, usamos o segundo valor retornado como uma variável descartável
area, _ = quad(f, 0, 2)

# Exibir a área calculada com duas casas decimais
print(f"Área sob a curva da função quadrática no intervalo [0, 2]: {area:.2f}")
```

### Algoritmos de Machine Learning e cálculo
Regressão linear e gradiente descendente

A regressão linear é um dos algoritmos mais simples de machine learning e tem como objetivo modelar a relação linear entre uma variável dependente e uma ou mais variáveis independentes. Para treinar o modelo, buscamos minimizar a função de custo, geralmente o erro quadrático médio (MSE).

O gradiente descendente é uma técnica de otimização que pode ser aplicada à regressão linear para encontrar os parâmetros do modelo que minimizam a função de custo. Ele atualiza os parâmetros do modelo iterativamente na direção oposta ao gradiente da função de custo.

Exemplo: Regressão linear com gradiente descendente

Vamos gerar um conjunto de dados sintético e aplicar a regressão linear com gradiente descendente para encontrar os parâmetros do modelo:

```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar a semente aleatória para reproduzir resultados
np.random.seed(0)

# Gerar dados sintéticos para regressão linear (relação y = 4 + 3X + ruído gaussiano)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Configurar os hiperparâmetros do gradiente descendente
eta = 0.1  # Taxa de aprendizado
n_iterations = 1000  # Número de iterações
m = len(X)  # Número de exemplos de treinamento

# Inicializar os parâmetros (coeficientes) da regressão linear aleatoriamente
theta = np.random.randn(2, 1)

# Adicionar uma coluna de 1's para representar o termo de interceptação (bias)
X_b = np.c_[np.ones((m, 1)), X]

# Executar o algoritmo de gradiente descendente
for iteration in range(n_iterations):
    # Calcular o gradiente da função de custo (erro quadrático médio)
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    
    # Atualizar os parâmetros usando o gradiente e a taxa de aprendizado
    theta -= eta * gradients

# Exibir os parâmetros encontrados pelo gradiente descendente
print("Parâmetros encontrados:", theta)

# Visualizar os dados originais e a linha de regressão ajustada pelo modelo
plt.scatter(X, y)  # Plotar os dados originais
plt.plot(X, X_b.dot(theta), color='red', label="Modelo")  # Plotar a linha de regressão
plt.xlabel("X")  # Adicionar rótulo ao eixo x
plt.ylabel("y")  # Adicionar rótulo ao eixo y
plt.legend()  # Adicionar legenda
plt.title("Regressão linear com gradiente descendente")  # Adicionar título ao gráfico
plt.grid()  # Adicionar grade ao gráfico
plt.show()  # Exibir o gráfico
```

Redes neurais e backpropagation

Redes neurais são algoritmos de machine learning baseados em neurônios artificiais, inspirados no funcionamento do cérebro humano. O treinamento de uma rede neural envolve ajustar os pesos e os bias dos neurônios para minimizar a função de custo.

SVM e otimização quadrática

As máquinas de vetores de suporte (SVM) são algoritmos de aprendizado supervisionado que podem ser usados para classificação e regressão. O objetivo do SVM é encontrar o hiperplano que melhor separa as classes (no caso da classificação) ou que melhor ajusta aos dados (no caso da regressão).

O treinamento de um SVM envolve a resolução de um problema de otimização quadrática, no qual buscamos maximizar a margem entre as classes ou minimizar o erro entre os dados e o hiperplano. A otimização quadrática é um campo do cálculo que estuda a otimização de funções quadráticas, geralmente com restrições lineares.

Exemplo: Classificação com SVM linear

Vamos utilizar o conjunto de dados "Iris" e aplicar a classificação com SVM linear usando a biblioteca Scikit-learn:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Carregar o conjunto de dados Iris
iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]  # Selecionar comprimento e largura da pétala
y = (iris["target"] == 2).astype(int)  # Classificar como Iris virginica (1) ou não (0)

# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalizar os dados (ajustar a média e o desvio padrão)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Treinar o classificador SVM linear
svm_clf = SVC(kernel="linear", C=1)
svm_clf.fit(X_train, y_train)

# Realizar previsões e calcular a acurácia
y_pred = svm_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Acurácia do SVM linear:", accuracy)

# Visualizar a fronteira de decisão do classificador SVM linear
x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

# Prever a classe para cada ponto na grade e redimensionar para o formato da grade
Z = svm_clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plotar a fronteira de decisão e os dados de treino
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, edgecolors='k')
plt.xlabel("Comprimento da pétala")
plt.ylabel("Largura da pétala")
plt.title("Classificação com SVM linear")
plt.show()
```

Exemplo: Implementação do gradiente descendente com Numpy

Vamos reescrever o exemplo de regressão linear com gradiente descendente utilizando as funções do Numpy para simplificar o código e torná-lo mais eficiente:

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerar dados sintéticos
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Configuração do gradiente descendente
eta = 0.1  # Taxa de aprendizado
n_iterations = 1000
m = len(X)

# Inicialização dos parâmetros
theta = np.random.randn(2, 1)

# Adicionar coluna de 1's para o bias/intercept
X_b = np.c_[np.ones((m, 1)), X]

# Gradiente descendente
for iteration in range(n_iterations):
    # Calcular o gradiente para os parâmetros atuais
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    
    # Atualizar os parâmetros usando o gradiente e a taxa de aprendizado
    theta -= eta * gradients

# Parâmetros encontrados
print("Parâmetros encontrados:", theta)

# Visualizar os dados e o modelo
plt.scatter(X, y)
plt.plot(X, X_b.dot(theta), color='red', label="Modelo")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Regressão linear com gradiente descendente e Numpy")
plt.grid()
plt.show()
```

### Implementação de Redes Neurais com Numpy
Neste tópico, veremos como implementar uma rede neural simples usando a biblioteca Numpy. Faremos a classificação do conjunto de dados "Iris" com uma rede neural de uma camada oculta.

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Função de ativação (sigmoid)
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Derivada da função de ativação
def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))

# Carregar o conjunto de dados Iris
iris = datasets.load_iris()
X = iris["data"]
y = iris["target"]

# Normalizar os dados
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Dividir em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Configuração da rede neural
input_size = X.shape[1]
hidden_size = 10
output_size = len(np.unique(y))
learning_rate = 0.01
epochs = 500

# Inicialização dos pesos e bias
W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros(hidden_size)
W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros(output_size)

# Codificar os rótulos como vetores one-hot
y_train_onehot = np.eye(output_size)[y_train]

# Treinar a rede neural
for epoch in range(epochs):
    # Forward pass: calcular a saída da rede neural
    z1 = np.dot(X_train, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)

    # Backward pass: calcular o erro e atualizar os pesos e bias
    delta2 = (a2 - y_train_onehot) * sigmoid_derivative(z2)
    dW2 = np.dot(a1.T, delta2)
    db2 = np.sum(delta2, axis=0)
    delta1 = np.dot(delta2, W2.T) * sigmoid_derivative(z1)
    dW1 = np.dot(X_train.T, delta1)
    db1 = np.sum(delta1, axis=0)

    # Atualização dos pesos e bias
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

# Testar a rede neural
z1_test = np.dot(X_test, W1) + b1
a1_test = sigmoid(z1_test)
z2_test = np.dot(a1_test, W2) + b2
a2_test = sigmoid(z2_test)

# Calcular a acurácia
y_pred = np.argmax(a2_test, axis=1)
accuracy = accuracy_score(y_test, y_pred)
print("Acurácia da rede neural:", accuracy)
```

Esta implementação básica de uma rede neural com Numpy ilustra como podemos utilizar a biblioteca para simplificar a criação e manipulação de matrizes e vetores em algoritmos de machine learning. Além disso, o conhecimento adquirido nesta implementação será útil para compreender como redes neurais mais complexas e profissionais, como as implementadas em bibliotecas como TensorFlow e PyTorch, funcionam internamente.

Embora esta seja uma implementação simples de uma rede neural, é possível modificar a arquitetura, função de ativação, taxa de aprendizado e outros hiperparâmetros para melhorar o desempenho e adaptá-la a diferentes problemas de classificação. No entanto, para problemas mais complexos e redes neurais mais profundas, é recomendável usar bibliotecas especializadas, como TensorFlow e PyTorch, que oferecem otimizações adicionais e melhor desempenho computacional.

### Comparação de desempenho entre Numpy e Python puro
Neste tópico, vamos comparar o desempenho entre Numpy e Python puro ao realizar operações matriciais. Isso ajudará a entender por que é benéfico usar Numpy em algoritmos de machine learning e como isso pode melhorar a eficiência computacional.

Exemplo: Multiplicação de matrizes

Vamos implementar a multiplicação de matrizes usando Python puro e Numpy, e comparar o tempo de execução entre os dois.

```python
import numpy as np
import random
import time

# Função para multiplicação de matrizes usando Python puro
def matrix_multiply_pure_python(A, B):
    # Inicializar a matriz resultado com zeros
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    
    # Iterar pelas linhas da matriz A
    for i in range(len(A)):
        # Iterar pelas colunas da matriz B
        for j in range(len(B[0])):
            # Iterar pelos elementos das linhas de A e colunas de B
            for k in range(len(B)):
                # Multiplicar os elementos correspondentes e acumular na matriz resultado
                result[i][j] += A[i][k] * B[k][j]
    return result

# Criar matrizes de teste com tamanho especificado
matrix_size = 200
A_python = [[random.random() for _ in range(matrix_size)] for _ in range(matrix_size)]
B_python = [[random.random() for _ in range(matrix_size)] for _ in range(matrix_size)]

# Converter as matrizes Python puro em matrizes Numpy
A_numpy = np.array(A_python)
B_numpy = np.array(B_python)

# Multiplicação de matrizes com Python puro e medir o tempo de execução
start_time = time.time()
result_python = matrix_multiply_pure_python(A_python, B_python)
end_time = time.time()
print("Tempo de execução com Python puro:", end_time - start_time)

# Multiplicação de matrizes com Numpy e medir o tempo de execução
start_time = time.time()
result_numpy = A_numpy.dot(B_numpy)
end_time = time.time()
print("Tempo de execução com Numpy:", end_time - start_time)
```

Ao executar este exemplo, você notará que a multiplicação de matrizes com Numpy é significativamente mais rápida do que com Python puro. Isso se deve ao fato de que Numpy é otimizado para operações matriciais e utiliza bibliotecas de álgebra linear de alto desempenho, como BLAS e LAPACK, para executar cálculos de forma eficiente.

Essa comparação demonstra o impacto que o uso de Numpy pode ter no desempenho dos algoritmos de machine learning. Ao tirar proveito das otimizações e recursos oferecidos pelo Numpy, é possível melhorar a eficiência computacional e reduzir o tempo de execução, especialmente em algoritmos que envolvem operações matriciais intensivas.

Exemplo: Cálculo da distância euclidiana

Vamos implementar o cálculo da distância euclidiana entre dois pontos usando operações vetorizadas e não vetorizadas com Numpy e comparar o tempo de execução entre os dois.

```python
import numpy as np
import time

# Função para calcular a distância euclidiana usando operações não vetorizadas
def euclidean_distance_non_vectorized(A, B):
    distance = 0
    # Iterar pelos elementos dos vetores A e B simultaneamente
    for a, b in zip(A, B):
        # Adicionar o quadrado da diferença entre os elementos correspondentes de A e B
        distance += (a - b) ** 2
    # Retornar a raiz quadrada da soma das diferenças ao quadrado
    return np.sqrt(distance)

# Função para calcular a distância euclidiana usando operações vetorizadas
def euclidean_distance_vectorized(A, B):
    # Utilizar a função np.linalg.norm para calcular a norma (distância euclidiana) entre os vetores A e B
    return np.linalg.norm(A - B)

# Criar vetores de teste com tamanho especificado
vector_size = 10000
A = np.random.rand(vector_size)
B = np.random.rand(vector_size)

# Calcular distância euclidiana com operações não vetorizadas e medir o tempo de execução
start_time = time.time()
result_non_vectorized = euclidean_distance_non_vectorized(A, B)
end_time = time.time()
print("Distância euclidiana (não vetorizada):", result_non_vectorized)
print("Tempo de execução com operações não vetorizadas:", end_time - start_time)

# Calcular distância euclidiana com operações vetorizadas e medir o tempo de execução
start_time = time.time()
result_vectorized = euclidean_distance_vectorized(A, B)
end_time = time.time()
print("Distância euclidiana (vetorizada):", result_vectorized)
print("Tempo de execução com operações vetorizadas:", end_time - start_time)
```

Ao executar este exemplo, você notará que o cálculo da distância euclidiana usando operações vetorizadas é significativamente mais rápido do que usando operações não vetorizadas. Isso ocorre porque as operações vetorizadas permitem que o Numpy tire proveito das otimizações de hardware, como as instruções SIMD (Single Instruction, Multiple Data), que permitem processar vários dados simultaneamente.

Essa comparação destaca a importância de usar operações vetorizadas ao trabalhar com Numpy e algoritmos de machine learning. Ao utilizar essas otimizações, você pode melhorar significativamente o desempenho e a eficiência computacional de seus algoritmos e reduzir o tempo de execução.
