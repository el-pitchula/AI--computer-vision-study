# 14 - Aplique o método do cotovelo em X e y para determinar qual a melhor quantidade de clusters para um algoritmo de clusterização 

```python
from sklearn.cluster import KMeans
```

```python
from IPython.display import clear_output
```

```python
sse = []
k_range = range(1, 10)
for k in k_range:
    km = KMeans(n_clusters = k)
    km.fit(X)
    sse.append(km.inertia_)
    clear_output(wait = True)
```

```python
sns.lineplot(x = k_range, y = sse)
```
